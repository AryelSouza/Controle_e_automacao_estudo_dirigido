```python
# ==========================================
# 2️⃣ Controlador P: ajuste de ganho
# ==========================================
K_values = [1, 5, 10, 20]

plt.figure()
for K in K_values:
    T = ctrl.feedback(K * G, 1)
    t, y = ctrl.step_response(T)
    plt.plot(t, y, label=f"K={K}")

plt.title("Controlador P — Efeito do ganho")
plt.xlabel("Tempo (s)")
plt.ylabel("Saída")
plt.legend()
plt.grid()
plt.show()
```

```python
# ==========================================
# 3️⃣ Controlador PI
# ==========================================
K = 5
Ti = 1
C_PI = K * ctrl.tf([Ti, 1], [Ti, 0])

T_pi = ctrl.feedback(C_PI * G, 1)
t, y = ctrl.step_response(T_pi)

plt.figure()
plt.plot(t, y)
plt.title("Controlador PI — Resposta ao Degrau")
plt.grid()
plt.show()
```

```python
# ==========================================
# 4️⃣ Controlador PD
# ==========================================
Td = 0.2
C_PD = K * ctrl.tf([Td, 1], [1])

T_pd = ctrl.feedback(C_PD * G, 1)
t, y = ctrl.step_response(T_pd)

plt.figure()
plt.plot(t, y)
plt.title("Controlador PD — Resposta ao Degrau")
plt.grid()
plt.show()
```

```python
# ==========================================
# 5️⃣ Compensador Lead
# ==========================================

z = 2
p = 8
C_lead = ctrl.tf([1, z], [1, p])

plt.figure()
ctrl.root_locus(C_lead * G, plot=True)
plt.title("Root Locus com Compensador Lead")
plt.grid()
plt.show()

# resposta
T_lead = ctrl.feedback(10 * C_lead * G, 1)
t, y = ctrl.step_response(T_lead)

plt.figure()
plt.plot(t, y)
plt.title("Resposta com Compensador Lead")
plt.grid()
plt.show()
```

```python
# ==========================================
# 6️⃣ Compensador Lag
# ==========================================

z = 0.2
p = 0.05
C_lag = ctrl.tf([1, z], [1, p])

T_lag = ctrl.feedback(5 * C_lag * G, 1)
t, y = ctrl.step_response(T_lag)

plt.figure()
plt.plot(t, y)
plt.title("Resposta com Compensador Lag")
plt.grid()
plt.show()
```
