```python
# Capítulo 4 - Simulações em Python
# Tema: Efeitos da Realimentação no Comportamento do Sistema
#instalando biblioteca de controle
!pip install control
import numpy as np
import matplotlib.pyplot as plt
import control as ctrl

# Sistema em malha aberta: G(s) = 1 / (s^2 + 3s + 2)
num = [1]
den = [1, 3, 2]
G = ctrl.tf(num, den)

# Controlador: C(s) = K
K = 5
C = ctrl.tf([K], [1])

# Realimentação unitária: H(s) = 1
H = ctrl.tf([1], [1])

# Função de transferência em malha fechada
T = ctrl.feedback(C*G, H)

# Resposta ao degrau
t, y = ctrl.step_response(T)

plt.figure()
plt.plot(t, y, label='Malha Fechada (K=5)')
plt.title('Resposta ao Degrau - Sistema com Realimentação')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Comparação com malha aberta
t2, y2 = ctrl.step_response(G)
plt.figure()
plt.plot(t2, y2, label='Malha Aberta')
plt.plot(t, y, label='Malha Fechada')
plt.title('Comparação entre Malha Aberta e Fechada')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()
```
